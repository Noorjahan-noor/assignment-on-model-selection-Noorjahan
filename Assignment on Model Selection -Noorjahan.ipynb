```python
for i in [ 'Age']:
    data[i]=data[i].fillna(data[i].median()) # replacing missing values using median value of the age column
```

```python
for i in ['Cabin', 'Embarked']:
    data[i]=data[i].fillna(data[i].mode()[0]) # replacing missing values using mode values of the cabin,embarked columns
```

```python
data.drop(['Cabin'],axis=1,inplace=True)
```

```python
data1=pd.get_dummies(data)
```

```python
data1.head()
```

```python
#  splitting the dataset into train,test
```

```python
from sklearn.model_selection import train_test_split # importing the library
```

```python
y=data1['Survived']#target
x=data1.drop('Survived',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

### KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
metric_k= []#empty list for appending accuracy of model for each k values
neighbors = np.arange(3, 15)#k value from 3 to 14

from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,accuracy_score

for k in neighbors:
  knn=KNeighborsClassifier(n_neighbors=k)
  model2=knn.fit(x_train,y_train)
  y_pred_knn=model2.predict(x_test)
  acc=accuracy_score(y_test,y_pred_knn)
  metric_k.append(acc)#appending accuracy of model for each k values
```

```python
metric_k
```

```python
plt.plot(neighbors,metric_k,'o-')#plotting the accuracy for each k value
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```

```python
knn=KNeighborsClassifier(n_neighbors=11)
model2=knn.fit(x_train,y_train)
y_pred_knn=model2.predict(x_test)  
#k=11 have more accuracy.so we can create model using this
```

```python
print(confusion_matrix(y_test,y_pred_knn))
```

```python
print(accuracy_score(y_test,y_pred_knn))
```

```python
 ### here we created a KNN Model with 74% accuracy
```

### K Fold Cross Validation

```python
from sklearn.model_selection import KFold # importing necessary library
```

```python
kfold_validator=KFold(10)
```

```python
for train_index, test_index in kfold_validator.split(x,y):
    print('Training Index: ',train_index)
    print('Testing Index: ',test_index)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_result=cross_val_score(model2,x,y,cv=kfold_validator)
```

```python
cv_result
```

```python
np.mean(cv_result)
```

### Stratified Cross Validation

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold_validator=StratifiedKFold(n_splits=10)
```

```python
for train_index, test_index in skfold_validator.split(x,y):
    print('Training Index: ',train_index)
    print('Testing Index: ',test_index)
```

```python
skcv_result=cross_val_score(model2,x,y,cv=skfold_validator)
```

```python
skcv_result
```

```python
np.mean(skcv_result)
```
